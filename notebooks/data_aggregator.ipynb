```python
import os
from datetime import datetime
from data_aggregator.binance_aggregator import aggregate_ohlc_data

# === Top 50 USDT pairs (spot Binance) ===
symbols = [
    "BTCUSDT", "ETHUSDT", "BNBUSDT", "XRPUSDT", "SOLUSDT", "ADAUSDT", "AVAXUSDT",
    "DOTUSDT", "DOGEUSDT", "LTCUSDT", "LINKUSDT", "UNIUSDT", "XLMUSDT",
    "TRXUSDT", "BCHUSDT", "ICPUSDT", "FILUSDT", "AAVEUSDT", "ATOMUSDT",
    "NEARUSDT", "FTMUSDT", "THETAUSDT", "HBARUSDT", "MKRUSDT", "GRTUSDT",
    "CAKEUSDT", "KSMUSDT", "FTTUSDT", "SANDUSDT", "CRVUSDT", "ENJUSDT", "ZECUSDT",
    "CELRUSDT", "CHZUSDT", "ZILUSDT", "ONTUSDT", "STORJUSDT", "MINAUSDT", "KAVAUSDT",
    "COMPUSDT", "ARUSDT", "ANKRUSDT", "BATUSDT", "EOSUSDT", "ALGOUSDT"
]

output_dir = "datasets/binance"
os.makedirs(output_dir, exist_ok=True)

for symbol in symbols:
    csv_path = os.path.join(output_dir, f"BINANCE_{symbol}_1H.csv")

    # Skip existing files
    if os.path.exists(csv_path):
        print(f"Skipping {symbol} — file already exists.")
        continue

    try:
        print(f"Aggregating data for {symbol}...")
        df = aggregate_ohlc_data(
            symbol=symbol,
            unit_of_time="hour",
            interval=1,
            start_date=None,
            end_date=datetime.now(),
        )

        # Validate and save
        if df is not None and not df.empty:
            df.to_csv(csv_path)
            print(f"✅ Data saved to {csv_path}")
        else:
            print(f"⚠️ No data returned for {symbol}, skipping.")

    except Exception as e:
        print(f"❌ Error aggregating {symbol}: {e}")

print("✅ All data aggregation complete.")
```
