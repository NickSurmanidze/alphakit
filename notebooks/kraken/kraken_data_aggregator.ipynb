```python
import os
from datetime import datetime
from data_aggregator.kraken_aggregator import aggregate_ohlc_data

# Major Kraken spot pairs with USD quote (using Kraken altname format)
# These correspond to the most liquid PF_ perpetual futures on Kraken Futures
pairs = [
    # Core crypto / USD
    "XBTUSD",   # BTC
    "ETHUSD",   # ETH
    "SOLUSD",   # SOL
    "XRPUSD",   # XRP
    "ADAUSD",   # ADA
    "DOTUSD",   # DOT
    "LINKUSD",  # LINK
    "LTCUSD",   # LTC
    "AVAXUSD",  # AVAX
    "ATOMUSD",  # ATOM
    "UNIUSD",   # UNI
    "FILUSD",   # FIL
    "AAVEUSD",  # AAVE
    "MKRUSD",   # MKR
    "COMPUSD",  # COMP
    "GRTUSD",   # GRT
    "KSMUSD",   # KSM
    "CRVUSD",   # CRV
    "ZECUSD",   # ZEC
    "XMRUSD",   # XMR
    "DASHUSD",  # DASH
    "EOSUSD",   # EOS
    "XLMUSD",   # XLM
    "TRXUSD",   # TRX
    "BCHUSD",   # BCH
    "NEARUSD",  # NEAR
    "ALGOUSD",  # ALGO
    "ICPUSD",   # ICP
    "SANDUSD",  # SAND
    "ENJUSD",   # ENJ
    "BATUSD",   # BAT
    "STORJUSD", # STORJ
    "KAVAUSD",  # KAVA
    "ANKRUSD",  # ANKR
    "OCEANUSD", # OCEAN
    "APEUSD",   # APE
    "GMTUSD",   # GMT
    "GALAUSD",  # GALA
    "FTMUSD",   # FTM
    "AXSUSD",   # AXS
    "MANAUSD",  # MANA
    "CHZUSD",   # CHZ
    "FLOWUSD",  # FLOW
    "SUIUSD",   # SUI
    "OPUSD",    # OP
    "ARBUSD",   # ARB
    "INJUSD",   # INJ
    "TIAUSD",   # TIA
    "PYTHUSD",  # PYTH
    "JUPUSD",   # JUP
]

output_dir = "datasets/kraken"
os.makedirs(output_dir, exist_ok=True)

for pair in pairs:
    csv_path = os.path.join(output_dir, f"KRAKEN_{pair}_1H.csv")

    if os.path.exists(csv_path):
        print(f"Skipping {pair} — file already exists.")
        continue

    try:
        print(f"Aggregating data for {pair}...")
        df = aggregate_ohlc_data(
            pair=pair,
            unit_of_time="hour",
            interval=1,
            start_date=None,
            end_date=datetime.now(),
        )

        if df is not None and not df.empty:
            df.to_csv(csv_path)
            print(f"✅ Saved {len(df):,} rows to {csv_path}")
        else:
            print(f"⚠️ No data returned for {pair}, skipping.")

    except Exception as e:
        print(f"❌ Error aggregating {pair}: {e}")

print("\n✅ All Kraken data aggregation complete.")
```
